```python
# ====== Install deps (run once) ======
!pip install reportlab
!pip install matplotlib
!pip install seaborn
!pip install scipy

# ====== Imports ======
import os, json, glob, math, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image, Table, TableStyle, PageBreak
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib import colors
from google.colab import files

# ====== CONFIG: adjust if your Drive structure differs ======
drive_base = "/content/drive/MyDrive/PhD_FraudDetection/experiments"
# Paths used earlier in notebook sessions
base_paths = {
    "fixed": {
        "lr": os.path.join(drive_base, "fixed_seed_runs/centralized_tradfi/logistic_regression"),
        "rf": os.path.join(drive_base, "fixed_seed_runs/centralized_tradfi/random_forest"),
        "xgb": os.path.join(drive_base, "fixed_seed_runs/centralized_tradfi/xgboost"),
    },
    "random": {
        "lr": os.path.join(drive_base, "random_runs/centralized_tradfi/logistic_regression"),
        "rf": os.path.join(drive_base, "random_runs/centralized_tradfi/random_forest"),
        "xgb": os.path.join(drive_base, "random_runs/centralized_tradfi/xgboost"),
    }
}

# Output
out_folder = os.path.join(drive_base, "summaries")
os.makedirs(out_folder, exist_ok=True)
pdf_path = os.path.join(out_folder, "Centralized_TradFi_Comparison_Report.pdf")
fig_folder = "/content/tmp_summary_figs"
os.makedirs(fig_folder, exist_ok=True)

# ====== Helper: load JSON results from a folder ======
def load_results_from_folder(folder):
    results = []
    if not os.path.isdir(folder):
        return results
    for path in sorted(glob.glob(os.path.join(folder, "*.json"))):
        try:
            with open(path, "r") as f:
                j = json.load(f)
            # normalize keys: some earlier scripts used 'f1' vs 'f1_score' etc.
            entry = {}
            entry['file'] = os.path.basename(path)
            # prefer available keys
            entry['accuracy']  = float(j.get('accuracy', j.get('acc', math.nan)))
            entry['precision'] = float(j.get('precision', math.nan))
            entry['recall']    = float(j.get('recall', math.nan))
            entry['f1']        = float(j.get('f1', j.get('f1_score', math.nan)))
            entry['auc']       = float(j.get('auc', j.get('auc_roc', math.nan)))
            entry['train_time']= float(j.get('train_time_seconds', j.get('train_time', math.nan)))
            entry['seed_or_run']= j.get('seed', j.get('seed_used', j.get('run', j.get('id', 'unknown'))))
            results.append(entry)
        except Exception as e:
            print("Failed to read", path, ":", e)
    return results

# ====== Aggregate per model & run-type ======
agg_rows = []
detailed = {}  # store lists of f1/precision/recall for plotting & stats

for run_type in ['fixed','random']:
    detailed[run_type] = {}
    for model_key, folder in base_paths[run_type].items():
        res = load_results_from_folder(folder)
        if len(res) == 0:
            print("WARNING: no results found for", run_type, model_key, "in", folder)
        df = pd.DataFrame(res)
        if df.empty:
            # create empty row
            agg_rows.append({
                "model": model_key,
                "run_type": run_type,
                "n_runs": 0,
                "accuracy_mean": np.nan,
                "accuracy_std": np.nan,
                "precision_mean": np.nan,
                "precision_std": np.nan,
                "recall_mean": np.nan,
                "recall_std": np.nan,
                "f1_mean": np.nan,
                "f1_std": np.nan,
                "auc_mean": np.nan,
                "auc_std": np.nan
            })
            detailed[run_type][model_key] = {"f1": [], "precision": [], "recall": []}
            continue

        # compute stats
        accuracy_mean, accuracy_std = df['accuracy'].mean(), df['accuracy'].std()
        precision_mean, precision_std = df['precision'].mean(), df['precision'].std()
        recall_mean, recall_std = df['recall'].mean(), df['recall'].std()
        f1_mean, f1_std = df['f1'].mean(), df['f1'].std()
        auc_mean, auc_std = df['auc'].mean(), df['auc'].std()

        agg_rows.append({
            "model": model_key,
            "run_type": run_type,
            "n_runs": len(df),
            "accuracy_mean": accuracy_mean,
            "accuracy_std": accuracy_std,
            "precision_mean": precision_mean,
            "precision_std": precision_std,
            "recall_mean": recall_mean,
            "recall_std": recall_std,
            "f1_mean": f1_mean,
            "f1_std": f1_std,
            "auc_mean": auc_mean,
            "auc_std": auc_std
        })
        detailed[run_type][model_key] = {
            "f1": df['f1'].dropna().tolist(),
            "precision": df['precision'].dropna().tolist(),
            "recall": df['recall'].dropna().tolist()
        }

# Master summary table (wide format)
master_df = pd.DataFrame(agg_rows)
# pretty metric columns with mean ± std
def mean_std_str(mean, std):
    if np.isnan(mean):
        return "NA"
    return f"{mean:.4f} ± {std:.4f}"
master_table_display = pd.DataFrame({
    "Model": master_df['model'] + " (" + master_df['run_type'] + ")",
    "Runs": master_df['n_runs'],
    "Accuracy": master_df.apply(lambda r: mean_std_str(r['accuracy_mean'], r['accuracy_std']), axis=1),
    "Precision": master_df.apply(lambda r: mean_std_str(r['precision_mean'], r['precision_std']), axis=1),
    "Recall": master_df.apply(lambda r: mean_std_str(r['recall_mean'], r['recall_std']), axis=1),
    "F1": master_df.apply(lambda r: mean_std_str(r['f1_mean'], r['f1_std']), axis=1),
    "AUC": master_df.apply(lambda r: mean_std_str(r['auc_mean'], r['auc_std']), axis=1)
})

# Save master table CSV for records
master_csv = os.path.join(out_folder, "Master_Summary_Table.csv")
master_table_display.to_csv(master_csv, index=False)

# ====== PLOTTING: boxplots for F1 across runs for each model (both fixed+random combined) ======
# Prepare a DataFrame with columns: model, run_type, f1
plot_rows = []
for run_type in detailed:
    for model_key in detailed[run_type]:
        for val in detailed[run_type][model_key]['f1']:
            plot_rows.append({"model": model_key, "run_type": run_type, "f1": val})
plot_df = pd.DataFrame(plot_rows)

plt.figure(figsize=(9,6))
sns.boxplot(x='model', y='f1', hue='run_type', data=plot_df)
plt.title("F1 distribution per model (fixed vs random runs)")
plt.tight_layout()
boxplot_path = os.path.join(fig_folder, "f1_boxplot_models.png")
plt.savefig(boxplot_path)
plt.close()

# Bar chart of mean F1 per model/run_type
mean_df = master_df.copy()
mean_df['label'] = mean_df['model'] + "_" + mean_df['run_type']
plt.figure(figsize=(8,5))
sns.barplot(x='label', y='f1_mean', data=mean_df)
plt.xticks(rotation=25)
plt.ylabel("Mean F1")
plt.title("Mean F1 per model and run-type")
plt.tight_layout()
barplot_path = os.path.join(fig_folder, "mean_f1_barplot.png")
plt.savefig(barplot_path)
plt.close()

# Boxplots for precision and recall as well
plt.figure(figsize=(9,6))
plot2 = []
for run_type in detailed:
    for model_key in detailed[run_type]:
        for val in detailed[run_type][model_key]['precision']:
            plot2.append({"model": model_key, "run_type": run_type, "precision": val})
plot2_df = pd.DataFrame(plot2)
if not plot2_df.empty:
    sns.boxplot(x='model', y='precision', hue='run_type', data=plot2_df)
    plt.title("Precision distribution per model")
    plt.tight_layout()
    precision_box = os.path.join(fig_folder, "precision_boxplot.png")
    plt.savefig(precision_box)
    plt.close()
else:
    precision_box = None

plt.figure(figsize=(9,6))
plot3 = []
for run_type in detailed:
    for model_key in detailed[run_type]:
        for val in detailed[run_type][model_key]['recall']:
            plot3.append({"model": model_key, "run_type": run_type, "recall": val})
plot3_df = pd.DataFrame(plot3)
if not plot3_df.empty:
    sns.boxplot(x='model', y='recall', hue='run_type', data=plot3_df)
    plt.title("Recall distribution per model")
    plt.tight_layout()
    recall_box = os.path.join(fig_folder, "recall_boxplot.png")
    plt.savefig(recall_box)
    plt.close()
else:
    recall_box = None

# ====== Variance / Stability Analysis ======
# For each model+run_type compute rel_std = std/mean*100% and decision stable/unstable using rule rel_std <= 3%
stability_rows = []
for _, row in master_df.iterrows():
    model = row['model']
    run_type = row['run_type']
    f1_mean = row['f1_mean']
    f1_std = row['f1_std']
    if np.isnan(f1_mean) or np.isnan(f1_std) or f1_mean == 0:
        rel_std = np.nan
        stable = "NA"
    else:
        rel_std = (f1_std / f1_mean) * 100
        stable = "Stable" if rel_std <= 3 else "Unstable"
    stability_rows.append({
        "model": model,
        "run_type": run_type,
        "f1_mean": f1_mean,
        "f1_std": f1_std,
        "f1_rel_std_percent": rel_std,
        "stability": stable,
        "n_runs": row['n_runs']
    })
stability_df = pd.DataFrame(stability_rows)

# ====== Cross-model statistical test (F1) ======
# Combine all F1 scores per model (from both run types) for comparison
combined_f1 = {}
for run_type in detailed:
    for model_key in detailed[run_type]:
        combined_f1.setdefault(model_key, [])
        combined_f1[model_key].extend(detailed[run_type][model_key]['f1'])

# Prepare test: if groups >=2, run normality & choose ANOVA or Kruskal-Wallis
stat_test_result = {}
groups = [combined_f1[m] for m in combined_f1 if len(combined_f1[m])>0]
group_names = [m for m in combined_f1 if len(combined_f1[m])>0]

if len(groups) >= 2:
    # Check normality with Shapiro (only if group size >=3)
    normal_flags = []
    for g in groups:
        if len(g) >= 3:
            stat, p = stats.shapiro(g)
            normal_flags.append(p > 0.05)
        else:
            normal_flags.append(False)
    all_normal = all(normal_flags) and len(normal_flags)>0
    if all_normal:
        # ANOVA
        f_stat, p_val = stats.f_oneway(*groups)
        stat_test_result['test'] = 'ANOVA'
        stat_test_result['f_stat'] = float(f_stat)
        stat_test_result['p_value'] = float(p_val)
    else:
        # Kruskal-Wallis
        h_stat, p_val = stats.kruskal(*groups)
        stat_test_result['test'] = 'Kruskal-Wallis'
        stat_test_result['h_stat'] = float(h_stat)
        stat_test_result['p_value'] = float(p_val)
else:
    stat_test_result['test'] = 'Not enough groups'
    stat_test_result['p_value'] = None

# ====== Create Combined PDF Report ======
styles = getSampleStyleSheet()
doc = SimpleDocTemplate(pdf_path, pagesize=A4)
story = []

# Title
story.append(Paragraph("<b>Centralized TradFi — Cross-Model Comparison Report</b>", styles['Title']))
story.append(Spacer(1,12))

intro = Paragraph(
    "This report summarizes the centralized TradFi experiments across Logistic Regression (LR), Random Forest (RF), and XGBoost (XGB). "
    "It includes a master summary table, boxplots of F1 distributions, a cross-model statistical analysis, and stability analysis (rel_std).",
    styles['BodyText']
)
story.append(intro)
story.append(Spacer(1,12))

# Master summary table (insert pandas table)
story.append(Paragraph("<b>Master Summary Table (mean ± std)</b>", styles['Heading2']))
story.append(Spacer(1,8))

# convert master_table_display to a reportlab table
mt_data = [list(master_table_display.columns)] + master_table_display.values.tolist()
mt_table = Table(mt_data, hAlign='LEFT', colWidths=[80,40,90,90,90,90,90])
mt_table.setStyle(TableStyle([
    ('GRID', (0,0), (-1,-1), 0.4, colors.black),
    ('BACKGROUND', (0,0), (-1,0), colors.lightgrey),
    ('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold'),
]))
story.append(mt_table)
story.append(PageBreak())

# Add F1 boxplot
story.append(Paragraph("<b>F1 Distribution per Model (boxplots)</b>", styles['Heading2']))
story.append(Spacer(1,8))
box_img = Image(boxplot_path, width=500, height=350)
story.append(box_img)
story.append(PageBreak())

# Add Mean F1 barplot
story.append(Paragraph("<b>Mean F1 per Model & Run-type</b>", styles['Heading2']))
story.append(Spacer(1,8))
bar_img = Image(barplot_path, width=450, height=300)
story.append(bar_img)
story.append(PageBreak())

# Precision / Recall boxplots if available
if precision_box:
    story.append(Paragraph("<b>Precision Distribution</b>", styles['Heading2']))
    story.append(Spacer(1,8))
    story.append(Image(precision_box, width=500, height=350))
    story.append(PageBreak())

if recall_box:
    story.append(Paragraph("<b>Recall Distribution</b>", styles['Heading2']))
    story.append(Spacer(1,8))
    story.append(Image(recall_box, width=500, height=350))
    story.append(PageBreak())

# Statistical test summary
story.append(Paragraph("<b>Cross-model Statistical Test (F1)</b>", styles['Heading2']))
story.append(Spacer(1,8))
if stat_test_result.get('test') == 'ANOVA':
    txt = f"ANOVA F={stat_test_result['f_stat']:.4f}, p={stat_test_result['p_value']:.4e}."
elif stat_test_result.get('test') == 'Kruskal-Wallis':
    txt = f"Kruskal-Wallis H={stat_test_result['h_stat']:.4f}, p={stat_test_result['p_value']:.4e}."
else:
    txt = "Not enough data to run cross-model statistical tests."
story.append(Paragraph(txt, styles['BodyText']))
story.append(Spacer(1,12))

# Stability analysis table
story.append(Paragraph("<b>Stability Analysis (F1 mean, std, rel_std %)</b>", styles['Heading2']))
story.append(Spacer(1,8))
stab_df = stability_df[['model','run_type','n_runs','f1_mean','f1_std','f1_rel_std_percent','stability']]
stab_data = [list(stab_df.columns)] + stab_df.values.tolist()
stab_table = Table(stab_data, hAlign='LEFT')
stab_table.setStyle(TableStyle([
    ('GRID', (0,0), (-1,-1), 0.4, colors.black),
    ('BACKGROUND', (0,0), (-1,0), colors.lightgrey),
    ('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold')
]))
story.append(stab_table)
story.append(PageBreak())

# Add short textual interpretation (auto-generated)
interp_lines = [
    "Automated interpretation (apply domain judgement):",
]
for _, r in stab_df.iterrows():
    rel = r['f1_rel_std_percent']
    stabtext = r['stability']
    interp_lines.append(f" - {r['model']} ({r['run_type']}): F1 mean={r['f1_mean']:.4f}, rel_std={rel:.2f}% → {stabtext}")

for line in interp_lines:
    story.append(Paragraph(line, styles['BodyText']))
    story.append(Spacer(1,6))

# build PDF
doc.build(story)

# Save CSVs for record
master_table_display.to_csv(os.path.join(out_folder, "Master_Summary_Table_readable.csv"), index=False)
stability_df.to_csv(os.path.join(out_folder, "Stability_Analysis.csv"), index=False)

print("Saved PDF:", pdf_path)
print("Saved master tables to:", out_folder)

# Download the PDF
files.download(pdf_path)
```

```python
# ===============================================================
# CENTRALIZED TRADFI — SEED 42 RE-RUN (ADVANCED VERSION)
# Logistic Regression, Random Forest, XGBoost
# Saves JSON + Confusion Matrix + Combined PDF
# ===============================================================

!pip install -q xgboost reportlab

import os, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image, PageBreak
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib import colors
from google.colab import files

# ===============================================================
# PATHS
# ===============================================================
BASE = "/content/drive/MyDrive/PhD_FraudDetection/experiments/centralized_tradfi"
os.makedirs(BASE, exist_ok=True)

FIXED_DIR = f"{BASE}/fixed_seed_runs"
os.makedirs(FIXED_DIR, exist_ok=True)

for m in ["lr","rf","xgb"]:
    os.makedirs(f"{FIXED_DIR}/{m}", exist_ok=True)

CM_DIR = "/content/tmp_cm_centralized"
os.makedirs(CM_DIR, exist_ok=True)

SEED = 42

# ===============================================================
# LOAD DATASET
# ===============================================================
assert "label" in df.columns, "Dataset missing label column."
X = df.drop("label", axis=1)
y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=y
)

# ===============================================================
# HELPER — SAVE CONFUSION MATRIX PLOT
# ===============================================================
def save_cm(cm, model_name):
    path = f"{CM_DIR}/{model_name}_cm_seed42.png"
    plt.figure(figsize=(4,4))
    plt.imshow(cm, cmap="Blues")
    plt.title(f"{model_name.upper()} (Seed 42)")
    plt.colorbar()
    for i in range(len(cm)):
        for j in range(len(cm[0])):
            plt.text(j, i, cm[i][j], ha="center", va="center", color="red")
    plt.tight_layout()
    plt.savefig(path)
    plt.close()
    return path

# ===============================================================
# HELPER — METRICS
# ===============================================================
def get_metrics(y_true, y_prob, y_pred):
    try:
        auc = roc_auc_score(y_true, y_prob)
    except:
        auc = float('nan')

    return {
        "accuracy": float(accuracy_score(y_true, y_pred)),
        "precision": float(precision_score(y_true, y_pred, zero_division=0)),
        "recall": float(recall_score(y_true, y_pred, zero_division=0)),
        "f1": float(f1_score(y_true, y_pred, zero_division=0)),
        "auc": float(auc),
        "confusion_matrix": confusion_matrix(y_true, y_pred).tolist()
    }


# ===============================================================
# MODEL 1 — LOGISTIC REGRESSION (scaled)
# ===============================================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

lr = LogisticRegression(max_iter=2000, solver="liblinear", random_state=SEED)
lr.fit(X_train_scaled, y_train)

y_prob_lr = lr.predict_proba(X_test_scaled)[:,1]
y_pred_lr = lr.predict(X_test_scaled)

lr_metrics = get_metrics(y_test, y_prob_lr, y_pred_lr)
lr_json_path = f"{FIXED_DIR}/lr/lr_seed_42.json"
with open(lr_json_path, "w") as f:
    json.dump(lr_metrics, f, indent=4)

lr_cm_path = save_cm(lr_metrics["confusion_matrix"], "lr")
print("LR done.")


# ===============================================================
# MODEL 2 — RANDOM FOREST (tuned)
# ===============================================================
rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    class_weight="balanced",
    random_state=SEED
)
rf.fit(X_train, y_train)

y_prob_rf = rf.predict_proba(X_test)[:,1]
y_pred_rf = rf.predict(X_test)

rf_metrics = get_metrics(y_test, y_prob_rf, y_pred_rf)
rf_json_path = f"{FIXED_DIR}/rf/rf_seed_42.json"
with open(rf_json_path, "w") as f:
    json.dump(rf_metrics, f, indent=4)

rf_cm_path = save_cm(rf_metrics["confusion_matrix"], "rf")
print("RF done.")


# ===============================================================
# MODEL 3 — XGBOOST (tuned)
# ===============================================================
pos = (y_train==1).sum()
neg = (y_train==0).sum()
scale_pos_weight = neg/pos if pos>0 else 1

xgb = XGBClassifier(
    n_estimators=350,
    max_depth=6,
    learning_rate=0.1,
    eval_metric="logloss",
    scale_pos_weight=scale_pos_weight,
    random_state=SEED,
    use_label_encoder=False
)
xgb.fit(X_train, y_train)

y_prob_xgb = xgb.predict_proba(X_test)[:,1]
y_pred_xgb = xgb.predict(X_test)

xgb_metrics = get_metrics(y_test, y_prob_xgb, y_pred_xgb)
xgb_json_path = f"{FIXED_DIR}/xgb/xgb_seed_42.json"
with open(xgb_json_path, "w") as f:
    json.dump(xgb_metrics, f, indent=4)

xgb_cm_path = save_cm(xgb_metrics["confusion_matrix"], "xgb")
print("XGB done.")


# ===============================================================
# BUILD COMBINED PDF
# ===============================================================
pdf_path = f"{BASE}/Centralized_TradFi_Seed42_Combined.pdf"
styles = getSampleStyleSheet()
doc = SimpleDocTemplate(pdf_path, pagesize=A4)
story = []

story.append(Paragraph("<b>Centralized TradFi — Seed 42 Combined Report</b>", styles["Title"]))
story.append(Spacer(1,15))

MODEL_DATA = {
    "Logistic Regression": (lr_metrics, lr_cm_path),
    "Random Forest": (rf_metrics, rf_cm_path),
    "XGBoost": (xgb_metrics, xgb_cm_path),
}

for model_name, (metrics, cm_path) in MODEL_DATA.items():
    story.append(Paragraph(f"<b>{model_name}</b>", styles["Heading1"]))
    story.append(Spacer(1,8))

    # table
    data = [
        ["Metric", "Value"],
        ["Accuracy", f"{metrics['accuracy']:.4f}"],
        ["Precision", f"{metrics['precision']:.4f}"],
        ["Recall", f"{metrics['recall']:.4f}"],
        ["F1 Score", f"{metrics['f1']:.4f}"],
        ["AUC", f"{metrics['auc']:.4f}"],
    ]

    table = Table(data)
    table.setStyle(TableStyle([
        ('GRID',(0,0),(-1,-1),0.5,colors.black),
        ('BACKGROUND',(0,0),(-1,0),colors.lightgrey),
    ]))
    story.append(table)
    story.append(Spacer(1,10))

    story.append(Paragraph("<b>Confusion Matrix</b>", styles["Heading2"]))
    story.append(Image(cm_path, width=400, height=300))
    story.append(PageBreak())

doc.build(story)
print("PDF GENERATED:", pdf_path)

try:
    files.download(pdf_path)
except:
    print("Download blocked, PDF saved to Drive.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
import json
from fpdf import FPDF
import os

# ============================================================
# LOAD DATASET
# ============================================================
df = pd.read_csv("/content/drive/MyDrive/PhD_FraudDetection/data/tradfi/final_tradfi_preprocessed.csv")

X = df.drop("label", axis=1)
y = df["label"]

SEED = 42
PDF_PATH = "/content/drive/MyDrive/PhD_FraudDetection/experiments/centralized_tradfi_seed42_combined_report.pdf"

SAVE_DIR = "/content/drive/MyDrive/PhD_FraudDetection/experiments/centralized_seed42_results"
os.makedirs(SAVE_DIR, exist_ok=True)

# ============================================================
# Train-test split
# ============================================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=SEED, stratify=y
)

# ============================================================
# Utility: Compute metrics
# ============================================================
def compute_metrics(y_true, y_pred, y_prob):
    return {
        "accuracy": round(accuracy_score(y_true, y_pred), 4),
        "precision": round(precision_score(y_true, y_pred, zero_division=0), 4),
        "recall": round(recall_score(y_true, y_pred, zero_division=0), 4),
        "f1": round(f1_score(y_true, y_pred, zero_division=0), 4),
        "auc": round(roc_auc_score(y_true, y_prob), 4),
    }

# ============================================================
# Utility: Save confusion matrix image
# ============================================================
def save_cm(y_true, y_pred, name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(4,3))
    plt.imshow(cm, cmap="Blues")
    plt.title(f"Confusion Matrix - {name}")
    plt.colorbar()
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    for i in range(2):
        for j in range(2):
            plt.text(j, i, cm[i,j], ha="center", va="center")
    out = f"{SAVE_DIR}/{name}_cm.png"
    plt.savefig(out, dpi=200, bbox_inches="tight")
    plt.close()
    return out

# ============================================================
# 1. LOGISTIC REGRESSION — Load existing JSON
# ============================================================
lr_json_path = "/content/drive/MyDrive/PhD_FraudDetection/experiments/fixed_seed_runs/centralized_tradfi/logistic_regression/seed_42.json"

with open(lr_json_path, "r") as f:
    lr_metrics = json.load(f)

# Recompute confusion matrix (need predictions)
lr_model = LogisticRegression(max_iter=500, random_state=SEED)
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_prob = lr_model.predict_proba(X_test)[:, 1]
lr_cm_path = save_cm(y_test, lr_pred, "LR_seed42")

# ============================================================
# 2. RANDOM FOREST — Re-run seed 42
# ============================================================
rf_model = RandomForestClassifier(n_estimators=300, random_state=SEED)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_prob = rf_model.predict_proba(X_test)[:, 1]
rf_metrics = compute_metrics(y_test, rf_pred, rf_prob)

with open(f"{SAVE_DIR}/rf_seed42.json", "w") as f:
    json.dump(rf_metrics, f, indent=4)

rf_cm_path = save_cm(y_test, rf_pred, "RF_seed42")

# ============================================================
# 3. XGBOOST — Re-run seed 42
# ============================================================
xgb_model = XGBClassifier(
    max_depth=5, learning_rate=0.1, n_estimators=300,
    subsample=0.9, colsample_bytree=0.9,
    random_state=SEED, eval_metric="logloss"
)
xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
xgb_prob = xgb_model.predict_proba(X_test)[:, 1]
xgb_metrics = compute_metrics(y_test, xgb_pred, xgb_prob)

with open(f"{SAVE_DIR}/xgb_seed42.json", "w") as f:
    json.dump(xgb_metrics, f, indent=4)

xgb_cm_path = save_cm(y_test, xgb_pred, "XGB_seed42")

# ============================================================
# GENERATE FINAL PROFESSIONAL PDF
# ============================================================
pdf = FPDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()

pdf.set_font("Arial", "B", 16)
pdf.cell(0, 10, "Centralized TradFi — Seed 42 Combined Report", ln=True)

def add_model_section(model_name, metrics, cm_path):
    pdf.set_font("Arial", "B", 14)
    pdf.cell(0, 10, f"\n{model_name}", ln=True)
    pdf.set_font("Arial", "", 12)

    for k,v in metrics.items():
        pdf.cell(0, 8, f"{k}: {v}", ln=True)

    pdf.ln(4)
    pdf.image(cm_path, w=120)
    pdf.ln(10)

add_model_section("Logistic Regression", lr_metrics, lr_cm_path)
add_model_section("Random Forest", rf_metrics, rf_cm_path)
add_model_section("XGBoost", xgb_metrics, xgb_cm_path)

pdf.output(PDF_PATH)
PDF_PATH
```
